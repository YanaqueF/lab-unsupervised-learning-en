# Challenge 4 - Data Clustering with K-Means

Count the values in `labels`.

```python
# Your code here:
kmeans_2 = KMeans(n_clusters=2, random_state=42)  # Initialize KMeans with 2 clusters
kmeans_2.fit(customers_scale)  # Fit the model
labels = kmeans_2.predict(customers_scale)  # Predict the cluster lab


df['labels'] = labels

print("\nDataset with Cluster Labels (2 Clusters):")
print(df.head())

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Grocery'], y=df['Milk'], hue=df['labels'], palette='coolwarm')
plt.title('Clusters of Customers (Grocery vs. Milk) - 2 Clusters')
plt.xlabel('Grocery')
plt.ylabel('Milk')
plt.legend(title='Cluster')
plt.show()

# Print the cluster assignments as a list
clusters = kmeans_2.labels_.tolist()
print("\nCluster assignments:")
print(clusters)
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 


# Initialize DBSCAN model with eps=0.5
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to the scaled data
dbscan.fit(customers_scale)

# Get the labels from DBSCAN and assign them to a new column in the DataFrame
df['labels_DBSCAN'] = dbscan.labels_

# Display the first few rows of the DataFrame with both K-Means and DBSCAN labels
print("\nDataset with Cluster Labels (K-Means and DBSCAN):")
print(df.head())

# Visualize the DBSCAN clusters in a 2D scatter plot (Grocery vs Milk)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Grocery'], y=df['Milk'], hue=df['labels_DBSCAN'], palette='coolwarm')
plt.title('Clusters of Customers (Grocery vs. Milk) - DBSCAN')
plt.xlabel('Grocery')
plt.ylabel('Milk')
plt.legend(title='Cluster')
plt.show()

# Compare K-Means and DBSCAN labels
print("\nComparison of K-Means and DBSCAN labels:")
comparison = df[['labels', 'labels_DBSCAN']]
print(comparison.value_counts())
```

Count the values in `labels_DBSCAN`.
